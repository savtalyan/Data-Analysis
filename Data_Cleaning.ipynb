```python
import pandas as pd

df = pd.read_csv(r"C:\Users\User\Downloads\movies.csv")
```

```python
# Let's set max rows display amount to 100, so we will be able to scroll

pd.set_option('display.max_rows', 100)
```

```python
df.head(10)
```

```python
# as we can see the same situation with unclean data is everywhere in this dataset.
# let's begin to clean it up
# First of all, I would like to remove parantheses form YEAR column

# And I think it would be better, to split YEAR column into two columns, Starting_Year and Ending_Year. 

df['YEAR'] = df['YEAR'].str.strip('()')
df['Starting_Year'] = df['YEAR'].str[:4]
df['Ending_Year'] = df['YEAR'].str[-5:]

# We get the dashes in the end of some values in Ending_Year, so I decided to replace with empty string everything except numbers
df['Ending_Year'] = df['Ending_Year'].str.replace(r'\D','', regex=True)

# Now it looks good. Lets drop YEAR column and change the indexes of two new columns to bring them forward
df.drop('YEAR', axis=1)
```

```python
df.head(100)
```

```python
df.columns
```

```python
# Let's drop columns we are not gonna use
columns = ['ONE-LINE', 'RunTime', 'Gross', 'STARS']
df.drop(columns, axis =1)
```

```python
df.columns
```

```python
# I want to change the order of columns

new_order = ['MOVIES', 'Starting_Year', 'Ending_Year', 'GENRE', 'RATING', 'VOTES']
df = df[new_order]
df.head(5)
```

```python
# Now it looks better 
# Let's go ahead and remove new line characters from GENRE column

df['GENRE'] = df['GENRE'].str.replace('\n', '')
df.head(100)
```

```python
# Now it is almost ready for our calculations. 
# We just need to set all NaN values equal to 0

df.fillna(0, inplace=True)
```
